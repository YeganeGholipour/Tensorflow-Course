```python
# Unzip the dataset

import zipfile
import os

with zipfile.ZipFile('/tmp/happy-or-sad.zip', mode='r') as archive:
    archive.extractall('/tmp/happy-or-sad-dir')
```

```python
# Make directories

happy_or_sad_dir = os.path.join('/tmp/happy-or-sad-dir')
happy_dir = os.path.join('/tmp/happy-or-sad-dir/happy')
sad_dir = os.path.join('/tmp/happy-or-sad-dir/sad')
```

```python
# Pics names

happy_name = os.listdir(happy_dir)
sad_name = os.listdir(sad_dir)

print(f"Name of the pics in happy directory {happy_name[0:10]}")
print(f"Name of the pics in sad directory {sad_name[0:10]}")
```

```python
# Plot the images

import matplotlib.pyplot as plt
import matplotlib.image as mpig

pic_index = 0

fig = plt.gcf()
fig.set_size_inches(16, 16)

happy_img_path = [os.path.join(happy_dir, img) for img in happy_name[pic_index:pic_index+8]]
sad_img_path = [os.path.join(sad_dir, img) for img in sad_name[pic_index:pic_index+8]]

for i, img_path in enumerate(happy_img_path+sad_img_path):
    subplot = plt.subplot(4,4,i+1)
    img = mpig.imread(img_path)
    plt.imshow(img)
    print(img.shape)
```

```python
# Number of pics

en_happy = len(happy_name)
len_sad = len(sad_name)
print(len_happy)
print(len_sad)
```

```python
# Building the convoloutional model

import tensorflow as tf

model = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape = (150, 150, 3)),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Flatten(), 
        tf.keras.layers.Dense(128, activation= 'relu'),
        tf.keras.layers.Dense(1, activation = 'sigmoid')

])

model.summary()
```

```python
# Adding imagegenerator for preprocessing and augmentation

from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
                rescale=.0/255
                #rotation_range = 40,
                #width_shift_range = 0.2,
                #height_shift_range = 0.2,
                #shear_range = 0.2,
                #zoom_range = 0.2,
                #horizontal_flip = True,
                #fill_mode = 'nearest'
)

train_generator = train_datagen.flow_from_directory(
                  happy_or_sad_dir, 
                  target_size = (150,150),
                  batch_size = 80,
                  class_mode = 'binary')

```

```python
# makeing callback class

class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if (logs.get('acc') > 0.999):
      print('\nDesired Accuracy is met, Stopping training...')
      self.model.stop_training = True

callbacks = myCallback()
```

```python
#from tensorflow.keras.optimizers import Adam
# Compiling the model

model.compile(
              loss = 'binary_crossentropy',
              optimizer = 'adam',
              metrics = ['acc'])

history = model.fit_generator(train_generator,
                              steps_per_epoch = 1,
                              epochs = 30,
                              verbose = 1,
                              callbacks = [callbacks]
)
```

```python
history.history.keys()
```

```python
# Plot the accuracy and loss of model

plt.plot(history.history['acc'])
plt.title('Training Accuracy')
plt.show()

plt.plot(history.history['loss'])
plt.title('Training Loss')
plt.show()
```
